# Essai pour la visualisation g√©ographiques avec Folium

```python
import folium
import json
import requests

# Cr√©ation de la carte
carteCannes = folium.Map(location=[43.551062, 7.013223],zoom_start=20)

# Affichage de la pop-up indiquant latitude et longitude lorsqu'on clique sur la carte
carteCannes.add_child(folium.LatLngPopup())

# Ajout de rep√®res avec des couleurs et logos personn√©lis√©s
folium.Marker(
    location=[43.5515, 7.0141],
    popup='Spot',
    icon=folium.Icon(icon="cloud")
).add_to(carteCannes)

folium.Marker(
    location=[43.5510, 7.0181],
    popup='Palais des Festivals',
    icon=folium.Icon(color='green')
).add_to(carteCannes)

folium.Marker(
    location=[43.5517, 7.0142],
    popup='un rep√®re au hasard',
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(carteCannes)

# Ajout d'un rep√®re avec affichage de graphe (utile pour l'√©volution de l'utilsation d'un collecteur)
stats = json.loads(requests.get("https://raw.githubusercontent.com/python-visualization/folium/master/examples/data/vis1.json").text)

folium.Marker(
    location=[43.5512, 7.0123],
    popup=folium.Popup(max_width=450).add_child(
        folium.Vega(stats, width=450, height=250))
).add_to(carteCannes)

# Ajout d'un cercle
folium.CircleMarker(
    location=[43.5514, 7.0126],
    radius=30,
    popup='H√¥tel de ville de Cannes',
    color='red',
    fill=True,
    fill_color='crimson'
).add_to(carteCannes)

carteCannes
```
